```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def quadratic(x, a=-1, b=2, c=10):
    return a * x**2 + b * x + c  # Ensure proper multiplication syntax for b and c

# Define the stages of the Waterfall Model
stages = ["Requirements", "Design", "Implementation", "Verification", "Maintenance"]
x_positions = np.linspace(-2, 2, len(stages))  # Equally spaced x positions for each stage
y_positions = quadratic(x_positions)  # Compute y positions using the quadratic function

# Generate quadratic curve for the background of the waterfall plot
x_curve = np.linspace(-2.5, 2.5, 500)
y_curve = quadratic(x_curve)

# Create the plot
plt.figure(figsize=(12, 6))

# Plot the quadratic curve as the backdrop (using a new color, e.g., darkgreen)
plt.plot(x_curve, y_curve, color="darkgreen", linestyle="--", label="Quadratic Curve")

# Add markers and vertical lines to represent the stages
for i in range(len(stages)):
    if i > 0:
        # Draw a horizontal segment to simulate the transition from the previous stage
        plt.plot(
            [x_positions[i-1], x_positions[i]],
            [y_positions[i-1], y_positions[i-1]],
            color="darkblue", linewidth=2
        )
        # Draw a vertical drop to simulate the waterfall effect between stages
        plt.plot(
            [x_positions[i], x_positions[i]],
            [y_positions[i-1], y_positions[i]],
            color="darkblue", linewidth=2
        )

    # Place markers and labels at each stage
    plt.scatter(x_positions[i], y_positions[i], color="red", zorder=5)
    plt.text(x_positions[i], y_positions[i] + 0.3, stages[i], ha="center", fontsize=12, color="black")

# Style the plot with titles and gridlines
plt.title("Waterfall Model Representation Using Quadratic Curve", fontsize=16, color="darkblue")
plt.xlabel("Stages of the Model", fontsize=14)
plt.ylabel("Value (Prediction)", fontsize=14)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")  # X-axis
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")  # Y-axis
plt.grid(color="gray", linestyle="--", linewidth=0.5, alpha=0.7)

# Add a legend to explain the plot components
plt.legend()

# Show the plot
plt.show()
```
